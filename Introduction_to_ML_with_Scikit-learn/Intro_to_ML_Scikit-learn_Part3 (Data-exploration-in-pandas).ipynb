### Renaming the classes


Basic Data Handling
-  The apply method offers a convenient way to manipulate pandas DataFrame entries along the column axis.
-  We can use a regular Python or lambda function as input to the apply method.
-  In this context, assume that our goal is to transform class labels from a string representation (e.g., "Iris-Setosa") to an integer representation (e.g., 0), which is a historical convention and a recommendation for compatibility with various machine learning tools.



```python
set(iris['species'])
```

```python
iris['species'] = iris['species'].apply(lambda x: 0 if x=='Iris-setosa' else x)
iris.head()
```

```python
iris.tail()
```

We have made use of the **lamba** function. These are just functions thata re defined in a simple way as already explained in functions: An expample:

`f = lambda x: sin(x) if x > 0 else 0 `


`.map vs. .apply`

-  If we want to map column values from one value to another, it is often more convenient to use the map method instead of apply.
-  The achieve the following with the apply method, we would have to call apply three times.



```python
classes = {'Iris-setosa': 0,
     'Iris-versicolor': 1,
     'Iris-virginica': 2}

iris['species'] = iris['species'].map(classes)
iris.head()
```

We can get the various category sum or averages

```python
categ = iris.groupby(['species']).sum()['sepal_length']
categ.head()
```

More sophisticated box and swam plots plots

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5), sharex=True)
for name, ax in zip(['petal_length', 'sepal_width'], axes):
    sns.boxplot(data=iris, x='species', y=name, ax=ax)
    sns.swarmplot(x="species", y=name, data=iris, palette="Set3", dodge=True, ax=ax)
    ax.set(ylabel= name, title=name)
```

### Correlation matrix
The correlation matrix gives us an idea of which rows are correlated extend to which they are correlated. It ranges from -1, to 1. With -1 strong negative correlation and +1 a strong positive correlation

```python
sns.set(font_scale=1.6)
plt.figure(figsize=(10,10))
corr_matrix = iris[column_names[:4]].corr()

cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr_matrix, annot=True, square=True, cmap='Blues')

plt.tight_layout()
```
